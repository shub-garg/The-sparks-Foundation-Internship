# City Wise Profit

```python
# city wise profit
profit_city = df.groupby("City").sum()['Profit']
profit_city
```

```python
profit_city.sort_values(ascending = False).head(10)
```

```python
profit_city.sort_values(ascending = False).tail(10)
```

```python
fig,axes = plt.subplots(1,2, figsize = (15,5))
profit_city.sort_values(ascending = False).head().plot(kind ='bar', ax = axes[0])
axes[0].tick_params(labelrotation = 10)
axes[0].set_title("Cities with Maximum Profit")

profit_city.sort_values(ascending = True).head().plot(kind='bar',ax = axes[1])
plt.xticks(rotation = 15)
plt.title("Cities with Minimum Profit")

plt.show()
```

# State wise Profit

```python
#STATE wise profit distributiom
state_profit = df.groupby('State').sum()['Profit']
```

```python
plt.figure(figsize =(15,8))
sns.barplot(x = state_profit.index, y=state_profit.values)
plt.xticks(rotation =50)
plt.title("STATE WISE PROFIT")
```

# Region wise Profit

```python
region_profit = pd.DataFrame(df.groupby('Region').sum()['Profit'])
```

```python
plt.pie(region_profit, labels = region_profit.index, autopct= "%.2f", explode = (0.05,0.05,0.05,0.20),radius=1.5)

centre_circle = plt.Circle((0,0), 0.70, fc = 'white')
fig= plt.gcf()
fig.gca().add_artist(centre_circle)
```

# Quantities Ordered Region Wise

```python
df_region=df.groupby('Region')['Quantity'].sum().reset_index()
print(df_region)

labels = df_region['Region'].unique()
plt.figure(figsize=(5,5))
plt.pie(df_region['Quantity'],autopct='%1.1f%%',explode=(0.02,0.02,0.02,0.02),)
plt.title('Quantities ordered by each region',size=13)
```

# Most Profitable Products

```python
dftop10_items = df.groupby('Sub-Category')['Profit'].sum().reset_index().sort_values(by='Profit',ascending=False)
dftop10_items.reset_index(drop=True,inplace=True)
dftop10_items=dftop10_items.head(10)
dftop10_items
```

```python
sns.catplot('Sub-Category','Profit',data=dftop10_items,kind='bar',aspect=1.5,height=6,palette='ch:2.5,-.2,dark=.3')
plt.title('Top 10 profitable products',size=10)
plt.xticks(size=10)
plt.yticks(size=10)
plt.ylabel('Cumulative profit',size=10)
plt.xlabel('Products',size=10)
```
